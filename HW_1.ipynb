```
params.results_dir = "results/"
SRA_list = params.SRA.split(",")

log.info ""
log.info " Q U A L I T Y   C O N T R O L  "
log.info "================================="
log.info "SRA number(s)         : ${SRA_list}"
log.info "Results location   : ${params.results_dir}"


process ReadSRA {
  publishDir "${params.results_dir}"

  input:
    val sra

  output:
    path "${sra}/*"

  script:
    """
    /content/sratoolkit.3.0.0-ubuntu64/bin/fasterq-dump ${sra} -O ${sra}/
    """ 
}

process FastQC {
  input:
    path x

  output:
    path "qc_result/*"

  script:
  """
    mkdir qc_result
    /content/FastQC/fastqc -o qc_result $x
  """
}

process MultiQC {
  publishDir "${params.results_dir}"

  input:
    path x

  output:
    path "multiqc_report.html"

  script:
  """
    multiqc $x
  """
}

workflow {
  data = Channel.of( SRA_list )
  ReadSRA(data)
  FastQC( ReadSRA.out )
  MultiQC( FastQC.out.collect() )
}
```
